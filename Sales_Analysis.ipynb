```python
# importing all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head(15)
```

```python
df.info()
```

```python
# drop unnecessary columns
df.drop(['Status','unnamed1'], axis =1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace =True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.describe()
```

```python
#Stats 
df[['Age','Orders','Amount']].describe()
```

## EDA - Exploratory Data Analysis

```python
fm = sns.countplot( x = 'Gender', data = df)
for bars in fm.containers:
    fm.bar_label(bars)
```

From the above graph we can see that most of the customers are females

```python
x = sns.countplot(data = df, x ='Age Group', hue = 'Gender')
for bars in x.containers:
    x.bar_label(bars)
```

From the above graph we can see that most of customers are between age 26 - 35 and most of them are Females

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x='Age Group', y ='Amount', data = sales_age)
```

```python
# total orders from 10 states
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (17,5)})
sns.barplot(data = sales_state, x= 'State', y = 'Orders')
```

```python
# total amount/sales from 10 states
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (17,5)})
sns.barplot(data = sales_state, x= 'State', y = 'Amount')
```

It is observed from the graphs that UP, Maharashtra and Karnataka have the most number of order and sales. 

```python
marriage_status = sns.countplot( x = 'Marital_Status', data = df)
sns.set(rc = {'figure.figsize' : (6,3)})
for bars in marriage_status.containers:
    marriage_status.bar_label(bars)
```

```python
# total orders from from married/unmarried customers
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(data = sales_state, x= 'Marital_Status', y = 'Amount', hue ="Gender")
```

We can clearly see that most of the buyers are married with females having the highest purchsing power

```python
occupation = sns.countplot( x = 'Occupation', data = df)
sns.set(rc = {'figure.figsize' : (20,5)})
for bars in occupation.containers:
    occupation.bar_label(bars)
```

```python
# total amount/sales from 10 states
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.barplot(data = sales_state, x= 'Occupation', y = 'Amount')
```

From the above grpah, we can clearly see that, most are the customers work in IT sector, follwed by healthcare and Aviation sector

```python
caterory_sales = sns.countplot( x = 'Product_Category', data = df)
sns.set(rc = {'figure.figsize' : (25,5)})
for bars in caterory_sales.containers:
    caterory_sales.bar_label(bars)
```

```python
# total amount/sales from 10 states
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.barplot(data = sales_state, x= 'Product_Category', y = 'Amount')
```

From the above graph, it is clearly visible that most of the products are sold from Clothing, Food and Electronics Category

```python
product_id = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.barplot(data = product_id, x= 'Product_ID', y = 'Orders')
```

## Conclusion :

Married women from age group 26-35 years from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food CLothing and Electronics Category.
